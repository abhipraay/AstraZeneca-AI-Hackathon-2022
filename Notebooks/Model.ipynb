```python
class PreProcess():
    
    def __init__(self,max_length=50):

        self.tokenizer = Tokenizer()
        self.max_length = max_length
    
    def fit_transform(self,X):
        f = open('./Data/Embeddings/embed.txt', encoding="utf8")
        embed={}
        for line in f.read().splitlines():
            word = line.split()[0]
            vec = list(map(float,line.split()[1:]))
            embed[word]=vec
        f.close()
        
        self.tokenizer.fit_on_texts(list(self.embed.keys()))
        
        X_enc = self.tokenizer.texts_to_sequences(X)
        padded_X = pad_sequences(X_enc,maxlen=self.max_length,padding ='post')
        return padded_X.tolist()
    
    def transform(self,X):
        X_enc = self.tokenizer.texts_to_sequences(X)
        padded_X = pad_sequences(X_enc,maxlen=self.max_length,padding ='post')
        return padded_X.tolist()
    
    def get_embed_wt(self):
        f = open('./Data/Embeddings/embed.txt', encoding="utf8")
        embed={}
        for line in f.read().splitlines():
            word = line.split()[0]
            vec = list(map(float,line.split()[1:]))
            embed[word]=vec
        f.close()
        
        embed_wt = np.zeros((vocab_size,200))
        for word,i in self.tokenizer.word_index.items():
            try:
                embed_wt[i]=embed[word]
            except:
                continue
            embed_wt[i]=embed[word]
        return embed_wt
    def get_vocab_size(self):
        return len(self.tokenizer.word_index)
    
    def fit(self,X):
        f = open('./Data/Embeddings/embed.txt', encoding="utf8")
        self.embed={}
        for line in f.read().splitlines():
            word = line.split()[0]
            vec = list(map(float,line.split()[1:]))
            self.embed[word]=vec
        f.close()
        
        self.tokenizer.fit_on_texts(list(self.embed.keys()))
        
```

```python
def read_chunks(chunck_numbers = range(49)):
    for i in chunck_numbers:
        file_name = './Data/Models/Chunks/' + str(i) + '.json'
        with open(file_name,'r') as f:
            yield load(f)
```

```python
import numpy as np
def decode(y):
    return np.argmax(y,1)

def score(y, ypred):
    return sum([1 for i in range(len(y)) if y[i]==ypred[i]])/len(y)
with open('./Example/Ex.txt' ,'r',encoding='utf8') as f:
    lines = f.readlines()
x_ex = []
y_ex = []
for line in lines:
    x_ex.append(line[1:])
    y_ex.append(int(line[0]))
x_ex = preprocess.transform(x_ex)
```

```python
from time import time
def score_model(model):
    t = time()
    p = decode(model.predict(x_ex))
    return (score(p,y_ex), time()-t)
```

```python
import keras_tuner as kt
```

```python
from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import GlobalMaxPooling1D
from keras.layers import Conv1D
from keras.layers import Dropout
from tensorflow.compat.v1.keras.layers import CuDNNLSTM
from keras.layers import Bidirectional
from keras.layers.convolutional import MaxPooling1D
def create_model(hp):
    conv_units = hp.Int('Conv units', min_value=100, max_value=600, step=100)
    conv_kernal_size =  hp.Choice('Conv kernal size', values=[3, 5])
    drop_out_rate = hp.Choice('Dropout rate', values=[0.1,0.2,0.3,0.4])
    dense_1 = hp.Choice('Dense-1', values=[32,64,128,256])
    dense_2 = hp.Choice('Dense-2', values=[32,64,128,256])
    model = Sequential()
    model.add(Embedding(vocab_size, 200, weights=[embed_wt], input_length=50, trainable=True))
    model.add(Conv1D(conv_units, 5, activation='relu'))
    model.add(Conv1D(conv_units, conv_kernal_size, activation='relu'))
    model.add(Conv1D(conv_units, 3, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Flatten())
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_1, activation='relu'))
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_1, activation='relu'))
    
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_2, activation='relu'))
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_2, activation='relu'))
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_2, activation='relu'))
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(dense_2, activation='relu'))
    model.add(Dropout(rate=drop_out_rate))
    model.add(Dense(32, activation='relu'))

    model.add(Dense(3, activation='sigmoid'))
    model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])
    return model
```
